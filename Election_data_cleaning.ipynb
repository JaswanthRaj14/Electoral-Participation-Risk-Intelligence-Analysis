```python
const_df = (
    df.groupby(
        ["state_name", "ac_name", "assembly_no", "ac_type"],
        as_index=False
    )
    .agg({
        "year": "first",
        "total_electors": "first",
        "total_votes": "first",
        "total_valid_votes": "first",
        "turnout_percentage": "first",
        "ac_total_candidates": "first",
        "margin": "max",  # winner margin
        "margin_percentage": "max"
    })
)
```

```python
const_df.duplicated(
    subset=["state_name", "ac_name", "assembly_no", "ac_type"]
).sum()
```

```python
const_df = const_df.sort_values(
    by=["state_name", "ac_name", "assembly_no"]
)
```

```python
const_df["turnout_change"] = (
    const_df.groupby(["state_name", "ac_name"])["turnout_percentage"]
    .diff()
)
```

```python
volatility_df = (
    const_df.groupby(["state_name", "ac_name"])["turnout_percentage"]
    .std()
    .reset_index()
    .rename(columns={"turnout_percentage": "turnout_volatility"})
)
```

```python
const_df = const_df.merge(
    volatility_df,
    on=["state_name", "ac_name"],
    how="left"
)
```

```python
const_df.head(10)
```

```python
print(const_df.columns)
```

```python
latest_df = (
    const_df.sort_values("assembly_no")
    .groupby(["state_name", "ac_name"])
    .tail(1)
)
```

```python
latest_df.head()
```

```python
latest_df["decline_flag"] = latest_df["turnout_change"] < 0
```

```python
state_vol_median = (
    latest_df.groupby("state_name")["turnout_volatility"]
    .transform("median")
)

latest_df["high_volatility_flag"] = (
    latest_df["turnout_volatility"] > state_vol_median
)
```

```python
state_turnout_avg = (
    latest_df.groupby("state_name")["turnout_percentage"]
    .transform("mean")
)

latest_df["low_turnout_flag"] = (
    latest_df["turnout_percentage"] < state_turnout_avg
)
```

```python
state_elector_median = (
    latest_df.groupby("state_name")["total_electors"]
    .transform("median")
)

latest_df["large_electorate_flag"] = (
    latest_df["total_electors"] > state_elector_median
)
```

```python
latest_df["decline_flag"] = latest_df["decline_flag"].astype(int)
latest_df["high_volatility_flag"] = latest_df["high_volatility_flag"].astype(int)
latest_df["low_turnout_flag"] = latest_df["low_turnout_flag"].astype(int)
latest_df["large_electorate_flag"] = latest_df["large_electorate_flag"].astype(int)
```

```python
latest_df["risk_score"] = (
    latest_df["decline_flag"] * 30 +
    latest_df["high_volatility_flag"] * 25 +
    latest_df["low_turnout_flag"] * 25 +
    latest_df["large_electorate_flag"] * 20
)
```

```python
priority_df = latest_df.sort_values(
    by="risk_score",
    ascending=False
)

priority_df.head(20)
```

```python
def calculate_slope(group):
    if len(group) < 2:
        return np.nan

    x = group["assembly_no"]
    y = group["turnout_percentage"]

    slope = np.polyfit(x, y, 1)[0]
    return slope

trend_df = (
    const_df.groupby(["state_name", "ac_name"])
    .apply(calculate_slope)
    .reset_index(name="turnout_trend_slope")
)
```

```python
latest_df = latest_df.merge(
    trend_df,
    on=["state_name", "ac_name"],
    how="left"
)
```

```python
latest_df["future_decline_flag"] = (
    latest_df["turnout_trend_slope"] < 0
).astype(int)
```

```python
latest_df["risk_score"] = (
    latest_df["risk_score"] +
    latest_df["future_decline_flag"] * 30
)
latest_df.head()
```

```python
latest_df.to_csv("election_final_dataset.csv", index=False)
```
